```python
import requests

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {
    "ingr": "champagne",  
    "limit": 10  }

headers = {
    "X-RapidAPI-Key": "7575919f42msh3911137aae4c362p1dd0afjsn604516d3e845",
    "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

df_champ = (response.json())
```

```python
df_champ
```

```python
# Fonction pour enregistrer les données dans un fichier CSV
def enreg_f(df, nom_f):
    with open(nom_f, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['foodId', 'label', 'category', 'foodContentsLabel', 'image']) # Specify the column names
        
        # Parcourir le dataframe et extraire les informations
        for item in df :
            food = df['parsed'][0]['food']
            food_id = food['foodId']
            label = food['label']
            category = food['category']
            food_contents_label = food['knownAs']
            image = food['image']
                        
            # Écrire les informations extraites dans une nouvelle ligne dans le fichier CSV
            writer.writerow([food_id, label, category, food_contents_label, image])

# Enregistrez les données dans un fichier CSV
enreg_f(df_champ,'df_c.csv')
```
